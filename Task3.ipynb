```python
import matplotlib.pyplot as plt

# Sample data: number of customers who purchased ("yes") each month
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
purchases = [20, 35, 40, 55, 50, 65]  # sample data showing increase in "yes" purchases

# Create line chart
plt.figure(figsize=(8, 5))
plt.plot(months, purchases, marker='o', color='blue', linestyle='-')
plt.title('Monthly Product Purchases (Yes responses)')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.grid(True)
plt.tight_layout()
plt.show()

```
